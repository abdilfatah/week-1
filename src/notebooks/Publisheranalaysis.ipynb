```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("raw_analyst_ratings.csv")
```

```python
df['domain'] = df['publisher'].apply(lambda x: x.split('@')[-1] if '@' in x else x)
```

```python
domain_counts = df['domain'].value_counts()
```

```python
domain_counts.head(10).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Publisher Domains by Contribution')
plt.xlabel('Domain')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['news_type'] = 'Unknown'
```

```python
top_domains = domain_counts.head(5).index
for domain in top_domains:
    publisher_news_type = df[df['domain'] == domain]['news_type'].value_counts()
    publisher_news_type.plot(kind='bar', color='salmon', edgecolor='black')
    plt.title(f'News Type Distribution for {domain}')
    plt.xlabel('News Type')
    plt.ylabel('Number of Articles')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```
